```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
```

## Style1: None

```python
# 이미지를 BGR 형식으로 읽어오기
# image = cv2.imread('온글잎 메리마체_하_120_none.png', cv2.IMREAD_UNCHANGED)
image = cv2.imread('온글잎 메리마체_하_120_none.png')
cv2_imshow(image)

# 이미지를 그레이스케일로 변환
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 가장자리 영역 검출
edges = cv2.Canny(gray, 100, 200)

# 가장자리 영역에서 픽셀 색상 추출
colors = []
for i in range(edges.shape[0]):
    for j in range(edges.shape[1]):
        if edges[i, j] != 0:  # 가장자리 픽셀
            colors.append(tuple(image[i, j]))

# 픽셀 색상별 등장 횟수 계산
color_counts = {}
for color in colors:
    if color in color_counts:
        color_counts[color] += 1
    else:
        color_counts[color] = 1

sorted_colors = sorted(color_counts.items(), key=lambda x: x[1], reverse=True)
print(sorted_colors)

# 가장 많이 등장하는 색상 찾기
# most_common_color = max(color_counts, key=color_counts.get)

# 결과 출력
print("텍스트 색상(BGR):", sorted_colors[1])
```

```python
import cv2
import numpy as np

# 이미지를 BGR 형식으로 읽어오기
image = cv2.imread('온글잎 메리마체_하_109_str.png')
cv2_imshow(image)

# 이미지를 그레이스케일로 변환
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)

# 이진화하여 텍스트 영역 분리
_, binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY_INV)
cv2_imshow(binary)

# 내부 영역 마스크 생성
internal_mask = cv2.bitwise_not(binary)

# 테두리 영역 마스크 생성
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
border_mask = np.zeros_like(gray)
cv2.drawContours(border_mask, contours, -1, 255, thickness=cv2.FILLED)

# 내부 영역에서 색상 추출
internal_colors = image[internal_mask != 0]
internal_color_counts = np.unique(internal_colors, axis=0, return_counts=True)
most_common_internal_color = internal_color_counts[0][np.argmax(internal_color_counts[1])]

# 테두리 영역에서 색상 추출
border_colors = image[border_mask != 0]
border_color_counts = np.unique(border_colors, axis=0, return_counts=True)
most_common_border_color = border_color_counts[0][np.argmax(border_color_counts[1])]

# 결과 출력
print("내부 영역에서 가장 많이 등장하는 색상:", most_common_internal_color)
print("테두리 영역에서 가장 많이 등장하는 색상:", most_common_border_color)
```

## 2. Stroke

```python
# str

# 이미지를 BGR 형식으로 읽어오기
image = cv2.imread('온글잎 메리마체_하_109_str.png', cv2.IMREAD_UNCHANGED)

cv2_imshow(image)

hist = []
print(image.shape)
# print(image.size / 3)

# 각 픽셀의 화소값
for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    hist.append(tuple(image[i, j]))

# 픽셀 색상별 등장 횟수 계산
color_counts = {}
for color in hist:
    if color in color_counts:
        color_counts[color] += 1
    else:
        color_counts[color] = 1

sorted_colors = sorted(color_counts.items(), key=lambda x: x[1], reverse=True)
print(sorted_colors)


image_colors = sorted_colors[1:3]
print(image_colors)


# 결과 출력
print("텍스트 색상(BGR):", image_colors[0])
print("테두리 색상(BGR):", image_colors[1])
```

어떤 색이 어떤 부분에 해당하는지 확인할 것
* 이미지 중앙에서 한칸씩 옆으로 갈 때, 가장 첫 번째로 마주치는 색(alpha 값이 255)이 str

```python
# 색의 순서 파악

half_height = image.shape[1] // 2
print(image_colors)

new_image_colors = []
for i in range(len(image_colors)):
  new_image_colors.append(image_colors[i][0])

print(new_image_colors)

detect_color = []
for i in range(image.shape[0]):
  pixel = tuple(image[i][half_height])

  if pixel[3] == 255:
    if pixel in new_image_colors:
      if pixel not in detect_color:
        detect_color.append(pixel)

print(detect_color)
```

```python
print("[Str] 테두리 색: ", detect_color[0])
print("[Str] 텍스트 색: ", detect_color[1])
```
