## Style1: None

```python
# 이미지를 BGR 형식으로 읽어오기
# image = cv2.imread('온글잎 메리마체_하_120_none.png', cv2.IMREAD_UNCHANGED)
image = cv2.imread('온글잎 메리마체_하_120_none.png')
cv2_imshow(image)

# 이미지를 그레이스케일로 변환
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 가장자리 영역 검출
edges = cv2.Canny(gray, 100, 200)

# 가장자리 영역에서 픽셀 색상 추출
colors = []
for i in range(edges.shape[0]):
    for j in range(edges.shape[1]):
        if edges[i, j] != 0:  # 가장자리 픽셀
            colors.append(tuple(image[i, j]))

# 픽셀 색상별 등장 횟수 계산
color_counts = {}
for color in colors:
    if color in color_counts:
        color_counts[color] += 1
    else:
        color_counts[color] = 1

sorted_colors = sorted(color_counts.items(), key=lambda x: x[1], reverse=True)
print(sorted_colors)

# 가장 많이 등장하는 색상 찾기
# most_common_color = max(color_counts, key=color_counts.get)

# 결과 출력
print("텍스트 색상(BGR):", sorted_colors[1])
```

## 1. None
1. 픽셀 RGBA의 빈도 수를 세고, 가장 많이 나오는 RGB값이 텍스트의 색이 됨
* 가장자리의 값은 픽셀이 흐리지만, RGB의 값은 변함이 없고 오직 alpha값만 변화함

```python
# none

# 히스토그램 추출하기
# (1) 원본 이미지
# (2) Gray Scale 이미지

# 이미지를 BGR 형식으로 읽어오기
image = cv2.imread('온글잎 메리마체_하_120_none.png', cv2.IMREAD_UNCHANGED)

cv2_imshow(image)

hist = []
print(image.shape)
# print(image.size / 3)

# 각 픽셀의 화소값
for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    hist.append(tuple(image[i, j]))

# 픽셀 색상별 등장 횟수 계산
color_counts = {}
for color in hist:
    if color in color_counts:
        color_counts[color] += 1
    else:
        color_counts[color] = 1

sorted_colors = sorted(color_counts.items(), key=lambda x: x[1], reverse=True)
print(sorted_colors)

image_colors = sorted_colors[1]

# 결과 출력
print(">> [None] 텍스트 색상(BGR):", image_colors)
```

* alpha 값이 255가 아닌 것은 자동으로 가장자리라는 의미가 됨. 중앙에 있는 값은 alpha값이 255임.
* alpha값이 0인 것은 무조건 배경임

```python
# 확인차원 - (배경 제외)한 개의 픽셀 값만 존재함
print(sorted_colors)
color_brg_cnt = {}
for color in hist:
  # print(color[0:3])
  if color[0:3] in color_brg_cnt and color[3] != 0:
      color_brg_cnt[color[0:3]] += 1
  elif (color[3] != 0):
      color_brg_cnt[color[0:3]] = 1

print(color_brg_cnt)
```

## 3. Glow
* 픽셀 수를 계산해 보았을 때, 가장 많은 픽셀의 빈도수는 텍스트 자체의 색임
* glow부분은 알파값이 존재(255가 아님) -> 이 구역을 구해서 색상 정보를 알아낼 예정

<과정>
1. 알파값이 존재하는 픽셀 위치를 탐색
2. 해당 구역 내에서 이웃 화소의 알파 값이 255인 구간 탐색(이 경우 해당 화소는 텍스트와 가장 붙어있다는 것을 의미함)
3. 색의 합과 픽셀의 개수를 계산하고, 평균을 계산

```python
# glow

# 이미지를 BGR 형식으로 읽어오기
image = cv2.imread('온글잎 메리마체_하_107_glow.png', cv2.IMREAD_UNCHANGED)

cv2_imshow(image)

hist = []
print(image.shape)
# print(image.size / 3)

# 각 픽셀의 화소값
for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    hist.append(tuple(image[i, j]))

# 픽셀 색상별 등장 횟수 계산
color_counts = {}
for color in hist:
    if color in color_counts:
        color_counts[color] += 1
    else:
        color_counts[color] = 1

sorted_colors = sorted(color_counts.items(), key=lambda x: x[1], reverse=True)
# sorted_colors = sorted(color_counts.items(), key=lambda x: x[0][3], reverse=True)   # 알파 값으로 정렬
print(sorted_colors)


image_colors = sorted_colors[1:2]   # 확실한 텍스트 색 하나만
print(image_colors)

# # 결과 출력 - 임시
# print("텍스트 색상(BGR):", sorted_colors[1])
# print("테두리 색상(BGR):", sorted_colors[2])
```

```python
# 알파 값을 채널로 빼서 이미지 출력

alpha_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(alpha_image)

print(alpha_image.shape)
print(alpha_image[100][100])

for i in range(image.shape[0]):
  for j in range(image.shape[1]):
    if 0 < image[i][j][3] < 255:
      alpha_image[i][j] = 255
    elif image[i][j][3] == 0:
      alpha_image[i][j] = 0
    else:
      alpha_image[i][j] = 150

cv2_imshow(alpha_image)
```

흰부분의 이웃 값의 알파 값이 255인 경우 픽셀 값을 더하기, 픽셀 개수 세서 평균 구하기
* 맨 가장자리의 경계도 구하기(영역 구하기)

```python
# 방법1 - 인접 이웃화소를 통한 색 추출
# 알파 값이 조금이라도 있는 경우 = 150
# 알파 값이 없는 경우 = 255

color_bgr = [0, 0, 0, 0]
pixel_cnt = 0

for i in range(1, alpha_image.shape[0] - 1):
  for j in range(1, alpha_image.shape[1] - 1):
    # 만일 번짐 영역이라면
    if alpha_image[i][j] == 255:
      # 텍스트 내부의 범위를 탐색(텍스트와 인접한 텍스트)
      if (alpha_image[i - 1][j - 1] == 150 or alpha_image[i + 1][j - 1] == 150 or alpha_image[i - 1][j + 1] == 150 or alpha_image[i + 1][j + 1] == 150):
        # print(image[i][j])
        color_bgr[0] += image[i][j][0]
        color_bgr[1] += image[i][j][1]
        color_bgr[2] += image[i][j][2]
        color_bgr[3] += image[i][j][3]
        pixel_cnt += 1

print(pixel_cnt)
print(color_bgr)

avg_color = [0, 0, 0, 0]

for i in range(len(color_bgr)):
  avg_color[i] = int(color_bgr[i] / pixel_cnt)

print(avg_color)
```

```python
# glow랑 동일한 방법

color_bgr = [0, 0, 0, 0]
pixel_cnt = 0

for i in range(1, alpha_image.shape[0] - 1):
  for j in range(1, alpha_image.shape[1] - 1):
    # 만일 번짐 영역이라면
    if alpha_image[i][j] == 255:
      # 텍스트 내부의 범위를 탐색(텍스트와 인접한 텍스트)
      if (alpha_image[i - 1][j - 1] == 150 or alpha_image[i + 1][j - 1] == 150 or alpha_image[i - 1][j + 1] == 150 or alpha_image[i + 1][j + 1] == 150):
        # print(image[i][j])
        color_bgr[0] += image[i][j][0]
        color_bgr[1] += image[i][j][1]
        color_bgr[2] += image[i][j][2]
        color_bgr[3] += image[i][j][3]
        pixel_cnt += 1

print(pixel_cnt)
print(color_bgr)

avg_color = [0, 0, 0, 0]

for i in range(len(color_bgr)):
  avg_color[i] = int(color_bgr[i] / pixel_cnt)

print(avg_color)    # 일단 이거라고 칠게
```

```python
# 알파값 영역 제거.. 가 아니라 고려 안하고 계산하면 되긴 하겠구나
# str코드 - 색의 순서 파악

half_height = image.shape[1] // 2
print(image_colors)

new_image_colors = []
for i in range(len(image_colors)):
  new_image_colors.append(image_colors[i][0])

print(new_image_colors)

detect_color = []
for i in range(image.shape[0]):
  pixel = tuple(image[i][half_height])

  if pixel[3] == 255:
    if pixel in new_image_colors:
      if pixel not in detect_color:
        detect_color.append(pixel)

print("먼저 감지되는 색의 순서: ", detect_color)
```

```python
print(">> [StrGlow] 번짐 색상(BGR):", avg_color)
print(">> [StrGlow] 테두리 색상(BGR):", detect_color[0])
print(">> [StrGlow] 텍스트 색상(BGR):", detect_color[1])
```
